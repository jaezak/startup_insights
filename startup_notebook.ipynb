```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from matplotlib import pyplot as plt
import numpy as np

# CSV file.
path = "..\startup_insights\startup_insights_csv.csv"

# Read the CSV file.
startup = pd.read_csv(path)
df = pd.DataFrame(startup)
```

```python
#There is a typo in row 43- "Finttech" should be "Fintech".
df['Industry'] = df['Industry'].replace('Finttech', 'Fintech')
```

```python
#Group data by industry types and visualize.
i_types = df.groupby("Industry")
industry_info = i_types["Industry"].count().sort_values(ascending=False)
types_graph = industry_info.plot(kind = "bar", color = "royalblue", figsize = (8,5), align = "center")
plt.title("Industry Types")
plt.xticks(rotation=45, ha="right")
plt.show()


```
